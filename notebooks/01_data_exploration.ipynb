```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the raw dataset
df = pd.read_csv('../data/raw/survey lung cancer.csv')
df.head()
```

```python
# Check for missing values and data types
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Check class imbalance
print(df['LUNG_CANCER'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='LUNG_CANCER', palette='Set2')
plt.title('Distribution of Lung Cancer Cases')
plt.show()
```

```python
# Visualize relationships (Correlation Heatmap)
# First, map target to numeric temporarily just for the heatmap
df_numeric = df.copy()
df_numeric['GENDER'] = df_numeric['GENDER'].map({'M': 1, 'F': 0})
df_numeric['LUNG_CANCER'] = df_numeric['LUNG_CANCER'].map({'YES': 1, 'NO': 0})

plt.figure(figsize=(14,10))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
